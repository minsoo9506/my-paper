```python
import os

PATH = './run_result_sim_error/'
os.listdir(PATH)
```

```python
err_001_01 = np.zeros((19918,500))
err_001_09 = np.zeros((19918,500))
err_010_01 = np.zeros((19918,500))
err_010_09 = np.zeros((19918,500))
```

```python
with open('./sim_data/smallTrainNoise_001_01.pickle', 'rb') as f:
    noise_001_01 = pickle.load(f)
with open('./sim_data/smallTrainNoise_001_09.pickle', 'rb') as f:
    noise_001_09 = pickle.load(f)
with open('./sim_data/smallTrainNoise_010_01.pickle', 'rb') as f:
    noise_010_01 = pickle.load(f)
with open('./sim_data/smallTrainNoise_010_09.pickle', 'rb') as f:
    noise_010_09 = pickle.load(f)
    
for data in [data for data in os.listdir(PATH) if 'smallTrainNoise_001_01' in data]:
    with open(PATH + data, 'rb') as f:
        tmp = pickle.load(f)
        err_001_01 += tmp / 4.0
    
for data in [data for data in os.listdir(PATH) if 'smallTrainNoise_001_09' in data]:
    with open(PATH + data, 'rb') as f:
        tmp = pickle.load(f)
        err_001_09 += tmp / 4.0
    
for data in [data for data in os.listdir(PATH) if 'smallTrainNoise_010_01' in data]:
    with open(PATH + data, 'rb') as f:
        tmp = pickle.load(f)
        err_010_01 += tmp / 4.0
    
for data in [data for data in os.listdir(PATH) if 'smallTrainNoise_010_09' in data]:
    with open(PATH + data, 'rb') as f:
        tmp = pickle.load(f)
        err_010_09 += tmp / 4.0
```

```python
def show_error(err, noise_idx):
    noise = []
    normal = []
    for i in np.arange(err.shape[0]):
        if i in noise_idx:
            noise.append(i)
        else:
            normal.append(i)

    noise_data = np.mean(err[noise, :], axis=0)
    normal_data = np.mean(err[normal, :], axis=0)
    return noise_data, normal_data
```

```python
noise_data, normal_data = show_error(err_001_01, noise_001_01)

fig, ax = plt.subplots(figsize=(7,7))
ax.plot(noise_data)
ax.plot(normal_data)
ax.legend(['noise', 'normal'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('epochs')
ax.set_ylabel('recon-error')
plt.title('recon-error changes')
```

```python
noise_data, normal_data = show_error(err_001_09, noise_001_09)

fig, ax = plt.subplots(figsize=(7,7))
ax.plot(noise_data)
ax.plot(normal_data)
ax.legend(['noise', 'normal'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('epochs')
ax.set_ylabel('recon-error')
plt.title('recon-error changes')
```
